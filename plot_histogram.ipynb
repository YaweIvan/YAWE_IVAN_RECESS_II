```python
import pandas as pd #type:ignore
import numpy as np #type:ignore
import matplotlib.pyplot as plt #type:ignore

#Load your data
data = pd.read_csv("data.csv")

# MaxPulse histogram
plt.hist(data['Maxpulse'], bins=10, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Histogram of Max Pulse')
plt.xlabel('MaxPulse')
plt.ylabel('Frequency')
plt.grid(True) #show grid lines for better readability
# Show the plot
plt.savefig("Maxpulse_chart.png")
plt.show()
```

```python
plt.hist(data['Calories'], bins=10, color='red', edgecolor='blue')
plt.title('Histogram of calories')
plt.xlabel('calories')
plt.ylabel('Frequency')
plt.grid(True) #show grid lines for better readability
# Show the plot 
plt.savefig("Calories_chart.png")
plt.show()
```
